# Segmentando os dados por renda e faixa de idade

## Importando os dados

```python
#importacao das bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#lendo dados
df=pd.read_csv('DadosDesafioCientista.csv')
```

```python
df['rendaMedia'] = pd.to_numeric(df['rendaMedia'], errors='coerce')
```

```python
df
```

```python
#separando os dados de sp
sp=df[df['estado']=='SP']
```

```python
sp
```

### Selecionando os bairros com renda acima de 9mil reais

Adotou-se que as rendas das classes A e B estão acima de 9 il reais. 

```python
sp1=sp[sp['rendaMedia']>9000]
```

```python
sp1.head()
```

## Organizando os dados de acordo com a faixa de idade

Primeiro criar uma coluna no dataframe com o total de população da faixa de idade entre 25 a 49 anos 

```python
#somando as colunas
sp1['popDe25a49'] = sp1['popDe25a34']+sp1['popDe35a49']
```

```python
sp1.head()
```

```python
#retirando as colunas desnecessárias
sp1.drop(['população','popAte9','popDe10a14','popDe15a19','popDe20a24','popDe25a34','popDe35a49','popDe50a59','popMaisDe60','domiciliosA1','domiciliosA2','domiciliosB1','domiciliosB2','domiciliosC1','domiciliosC2','domiciliosD','domiciliosE','faturamento','potencial'],axis=1, inplace=True)
```

Colocando os bairros em ordem decrescente de população da faixa de 25 a 49 anos

```python
sp2=sp1['popDe25a49'].sort_values(ascending=False)
```

```python
sp2
```
